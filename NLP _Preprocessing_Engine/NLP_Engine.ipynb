# Task 1: Conceptual Understanding

### 1. Difference between "Love" and "love"
"Love" and "love" are treated as different words in case-sensitive NLP systems. 
After preprocessing (lowercasing), both become "love" and are treated the same.

### 2. What happens if stopwords are not removed?
If stopwords are not removed, the dataset will contain unnecessary words like "is", "the", "and". 
This increases noise and reduces model performance.

### 3. When removing stopwords can be harmful
1. Sentiment Analysis: "not happy" becomes "happy" (wrong meaning)
2. Question Answering: Removing words affects sentence meaning

### 4. Stemming vs Lemmatization
Stemming cuts words to root form (running → runn) but may be incorrect.
Lemmatization converts words properly (running → run).

### IMPORTS

```python
import re
from collections import Counter
```

# Task 2: Preprocessing Function

```python
def preprocess_text(text):
    # Lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove special characters
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short words (except no, not)
    tokens = [w for w in tokens if len(w) > 2 or w in ['no', 'not']]

    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence
```

# Task 3: Stress Testing

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooove this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

TESTING OUTPUT

```python
results = []

for s in sentences:
    tokens, clean = preprocess_text(s)
    results.append((s, tokens, clean))

    print("\nOriginal:", s)
    print("Tokens:", tokens)
    print("Clean Sentence:", clean)
```

# Task 4: Token Analysis

```python
for s, tokens, clean in results:
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(word) for word in tokens) / total_tokens if total_tokens > 0 else 0

    print("\nSentence:", clean)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_length, 2))
```

# Task 5: Frequency Analysis

```python
all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nLeast 5 Frequent Words:")
print(counter.most_common()[-5:])
```

# Task 6: Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.append(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

output = full_pipeline(sentences)
print(output)
```
