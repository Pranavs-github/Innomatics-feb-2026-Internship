##  Problem 1: Mobile Recharge Validation System
 Problem Statement
Create a function to validate a mobile recharge request.
Rules:
Recharge amount must be ≥ ₹50
Valid plans: 199, 299, 399, 599
User should be allowed to retry if invalid


Real-Time Use
Telecom recharge apps
Payment gateways

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        try:
            amount = int(input("Enter recharge amount: "))

            if amount < 50:
                print("Recharge amount must be at least ₹50.")
            elif amount not in valid_plans:
                print("Invalid plan. Valid plans are:", valid_plans)
            else:
                print(f"Recharge successful for ₹{amount}")
                break

            retry = input("Do you want to try again? (yes/no): ").lower()
            if retry != "yes":
                print("Exiting recharge process.")
                break

        except ValueError:
            print("Invalid input. Please enter a number.")
validate_recharge()
```

## Problem 2: Inventory Reorder Alert System
Problem Statement
Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15
Print Stock OK otherwise

Real-Time Use
Warehouse systems
Retail management


```python
products = {
    "Apples": 20,
    "Bananas": 10,
    "Oranges": 25,
    "Mangoes": 12
}

for product, stock in products.items():
    if stock < 15:
        print(f"{product}: Reorder Alert")
    else:
        print(f"{product}: Stock OK")
```

## Problem 3: Student Result Processing System
Problem Statement
Create a function that:
Accepts a list of student marks


Calculates average marks
Prints Pass if average ≥ 50 else Fail

Real-Time Use
Online exam systems
Evaluation portals


```python
def student_result():
    li = []
    for i in range(10):
        marks = int(input("Enter the marks of the student: "))
        li.append(marks)
    print(li)
    sum = 0
    for i in range(len(li)):
        sum = sum + li[i]
    average = sum /len(li)
    if average >= 50:
        print("Student Passed.")
    else:
        print("Student Failed.")
        
student_result()
```

## Problem 4: Cab Fare Estimator with Retry Option
Problem Statement
Create a cab fare calculator:
Base fare = ₹50
₹12 per km
Peak hour → 25% extra
Allow retry using while loop

Real-Time Use
Ride-hailing apps
Transport systems


```python
def cab_fare_estimator(trip, peak_hour):
    base_fare = 50
    distance_fare = 12 * trip
    total = base_fare + distance_fare
    
    if peak_hour == "yes":
        total = total + (0.25 * total)  # add 25% extra
    
    return total


while True:
    distance_travelled = float(input("Enter distance travelled (in km): "))
    time = input("Is it peak hour (yes/no): ").lower()
    
    charge = cab_fare_estimator(distance_travelled, time)
    print(f"Total Fare: ₹{charge:.2f}")
    
    retry = input("Do you want to calculate again (yes/no): ").lower()
    if retry != "yes":
        break
```

## Problem 5: Employee Attendance Eligibility Checker
Problem Statement
Create a function that:
Accepts attendance list ("P", "A")
Calculates attendance percentage
Returns Eligible if ≥ 75% else Not Eligible

Real-Time Use
HR payroll systems
Internship tracking


```python
def eligibility_checker(l1):
    print(l1)
    count = 0
    for i in range(len(l1)):
        if l1[i] == 'P':
            count = count + 1
        else:
            continue
    total_days = len(l1)
    percentage = (count/total_days) * 100
    if percentage >= 75:
        print(f"the attdence percentage is: {percentage}")
        return "Employee Eligible"
    else:
        print(f"the attdence percentage is: {percentage}")
        return "Employee Not Eligible"    
list1 = ['P', 'P', 'A', 'A', 'P', 'A', 'P', 'A', 'P']
print(eligibility_checker(list1))
```

## Problem 6: Password Strength Checker
Problem Statement
Create a function to check password strength.

Rules:
Minimum length = 8
Must contain at least one digit
Must contain at least one special character (@#$)


Real-Time Use
Login systems
Security validation


```python
def password_strength(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"
    
    if len(password) >= 8:
        for ch in password:
            if ch.isdigit():
                has_digit = True
            if ch in special_chars:
                has_special = True
        return has_special and has_digit
    
    

check_pass = input()
print("Password contains digit and special character: ", password_strength(check_pass))
```
