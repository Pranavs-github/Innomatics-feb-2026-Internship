# Problem 1: Unique Words in a Sentence
Problem Statement
Given a sentence, identify all unique words and display:
The count of unique words
The unique words themselves

Input
Enter sentence: python is easy and python is powerful

Expected Output
Unique words count: 5  
Unique words: {'python', 'is', 'easy', 'and', 'powerful'}


```python
str1 = input()
li = str1.split()
print(li)
s1 = set(li)
print(s1)
print(len(s1))
```

# Problem 2: Highest Salary from Employee Data
Problem Statement
Store employee names and their salaries, and determine which employee has the highest salary.

Input Format
A dictionary where:
Keys represent employee names
Values represent employee salaries
Example Input
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

Expected Output
Highest Salary: Ravi - 75000


```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
items = employees.items()
highest = max(items, key=lambda pair: pair[1])
print("Highest Salary:", highest[0], "-", highest[1])
```

# Problem 3: Find Maximum and Minimum Values
Problem Statement
Find the maximum and minimum values from a list of numbers.

Input Format
A list of integers where each element represents a numeric value.
Example Input
numbers = [45, 22, 89, 10, 66]

Expected Output
List: [45, 22, 89, 10, 66]  
Max: 89  
Min: 10


```python
numbers = [45, 22, 89, 10, 66]
max = numbers[0]
min = numbers[0]
n = len(numbers)
for i in range(n):
    if numbers[i] > max:
        max = numbers[i]
for i in range(n):
    if numbers[i] < min:
        min = numbers[i]
print("max:", max)
print("min:", min)
```

# Problem 4: Count Products Above a Price Threshold
Problem Statement
Count how many products have a price greater than 1000.

Input Format
A list of integers where each element represents the price of a product.
Example Input
prices = [450, 1200, 899, 1500, 300]

Expected Output
Products above 1000: 2



```python
prices = [450, 1200, 899, 1500, 300]
n = len(prices)
count = 0
for i in range(n):
    if prices[i] > 1000:
        count = count + 1
print(f"product above 1000: {count}")
```

# Problem 5: Calculate Attendance Percentage
Problem Statement
Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).

Example Input
attendance = ["P", "P", "A", "P", "P"]

Expected Output
Attendance Percentage: 80.0


```python
attendance = ["P", "P", "A", "P", "P"]
n = len(attendance)
present_count = 0
for i in range(n):
    if attendance[i] == "P":
        present_count = present_count + 1
Attendance_Persentage = (present_count/n) * 100
print(f"Attendance Persentage: {Attendance_Persentage}")
```

# Problem 7: Count Character Frequency
Problem Statement
Count how many times each character appears in a given string.

Expected Output
{'p': 2, 'y': 1, 't': 1, 'h': 1, 'o': 1, 'n': 1}


```python
from collections import Counter

text = "I want to sleep."
char_count = dict(Counter(text))
print(char_count)
```

```python
text = "This is another way to solve"

char_count = {}

for ch in text:
    if ch in char_count:
        char_count[ch] = char_count[ch] + 1
    else:
        char_count[ch] = 1
print(char_count)
```

# Problem 10: Calculate Average Marks
Problem Statement
Calculate the average marks of a student from a list of marks.

Expected Output
Average Marks: 72.5


```python
marks = [100,92, 89,78,94,67,48,91]
sum = 0
for i in range(len(marks)):
    sum = sum + marks[i]
average_marks = sum / len(marks)
print(average_marks)
```
