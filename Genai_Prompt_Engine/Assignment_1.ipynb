## Task 2: Multi-Input Prompt System

Creating a prompt template with multiple inputs: topic, audience, and tone.

```python
from langchain_core.prompts import PromptTemplate

template_multi = "Explain {topic} for {audience} in a {tone} tone"

prompt_multi = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template_multi
)

print(prompt_multi.format(topic="AI", audience="beginners", tone="friendly"))
print(prompt_multi.format(topic="Python", audience="kids", tone="fun"))
print(prompt_multi.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

## Task 3: Prompt Variations Engine

Creating multiple prompt templates for different styles: teaching, interview, storytelling.

```python
from langchain_core.prompts import PromptTemplate

# Teaching style
teaching_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Interview style
interview_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

# Storytelling style
story_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print(teaching_prompt.format(topic=topic))
print(interview_prompt.format(topic=topic))
print(story_prompt.format(topic=topic))
```

## Task 4: ChatPromptTemplate System

Using system and user roles to create structured chat prompts.

```python
from langchain_core.prompts import ChatPromptTemplate

chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {role}"),
    ("user", "Explain {topic}")
])

messages = chat_template.format_messages(
    role="teacher",
    topic="Neural Networks"
)

print(messages)
```

## Task 5: Input Validation

Validating audience and tone inputs.

```python
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        audience = "beginner"

    if tone not in valid_tone:
        tone = "formal"

    return audience, tone
```

## Task 6: Prompt Generator

Generating dynamic prompts based on topic, audience, tone, and style.

```python
from langchain_core.prompts import PromptTemplate

def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    styles = {
        "teaching": "Explain {topic} step by step for {audience} in a {tone} tone",
        "interview": "Generate interview questions on {topic} for {audience} in a {tone} tone",
        "storytelling": "Explain {topic} as a story for {audience} in a {tone} tone"
    }

    template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=styles[style]
    )

    return template.format(topic=topic, audience=audience, tone=tone)


print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

## Task 7: Template Reusability Test

Using the same function with different inputs.

```python
topics = ["AI", "Python", "Data Science", "Machine Learning", "Deep Learning"]

for t in topics:
    print(generate_prompt(t, "beginner", "casual", "teaching"))
```
