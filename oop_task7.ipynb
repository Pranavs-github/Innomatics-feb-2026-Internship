### Problem 1: Smart Light Controller
Problem Statement
Create a class to control a smart light that can be turned ON or OFF and display its status.

üåç Real-Time Use
Smart home automation systems

üì• Sample Input
Light Name: Bedroom Light
Action: ON

üì§ Expected Output

```python
class Home:
    def __init__(self, name, action):
        self.name = name
        self.action = action
    def show_status(self):
        print(self.name + " is " + self.action)
light = Home("panta", "off")
light.show_status()
    
```

```python
class Home:
    def __init__(self, name):
        self.name = name
        self.action = "oFF" #default
    def turn_on(self):
        self.action = "ON"
        print(self.name + " is " + self.action)
    def turn_off(self):
        self.action = "OFF"
        print(self.name + " is " + self.action)
light = Home("koti")
light.turn_on()
light.turn_off()
```

### Problem 2: Employee ID Card System
 Problem Statement
Create a class that stores employee name, ID, and department and displays ID card details.

Real-Time Use
Corporate HR systems

üì• Sample Input
Employee Name: Rahul
Employee ID: EMP102
Department: IT

üì§ Expected Output
Employee ID Card
Name: Rahul
ID: EMP102
Department: IT


```python
class Employee:
    def __init__(self, name, id, dept):
        self.name = name
        self.id = id
        self.dept = dept
        
    def show_details(self):
        print("Employee ID card")
        print("Name:", self.name)
        print("ID:", self.id)
        print("Department:", self.dept)
        
emp = Employee("Rahul" , "EMP123", "Sales")
emp.show_details()
```

### Problem 4: Product Price Tag Generator

Create a class that stores product name and price and prints a formatted price tag.

 Real-Time Use
Retail billing systems

üì• Sample Input

Price: 2499

üì§ Expected Output

Price: ‚Çπ2499


```python
class Product:
    def __init__(self, product, price):
        self.product = product
        self.price = price
    def price_tag(self):
        print("Product Price Tag")
        print("Product:", self.product)
        print("Price:", self.price)
        
item = Product("pen", 10)
item.price_tag()
```

### Problem 5: Movie Rating Display System
Problem Statement
Create a class that stores movie name and rating and displays movie details.

Real-Time Use
Streaming platforms

üì• Sample Input
Movie Name: Inception
Rating: 4.8

üì§ Expected Output
Movie: Inception
Rating: 4.8 / 5


```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    def movie_details(self):
        print("Movie Details")
        print("Movie:", self.name)
        print("Rating:", str(self.rating) + " / 5")
movie_title = Movie("xzy", 4.5)
movie_title.movie_details()
```
